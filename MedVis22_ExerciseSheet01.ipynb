# Example Case in Python



Here is some sample code how medical image data can be read and displayed using Python. The dataset is from Jeff Mather (DICOM Example Files [link text](https://de.mathworks.com/matlabcentral/fileexchange/2762-dicom-example-files), MATLAB Central File Exchange. Retrieved April 8, 2022.).

As you may know, medical imaging like CT and MRI often produce volume data by generating cross-sectional slices through the body. One way to view this data is by visualising each slice. In our DICOM data set each slice is saved as a separate file. Therefore, you first have to traverse the slice files to read and then save them in an 3D array. This is done by the function read_dcm_volume. You can then choose one slice of this volume by entering its index, e.g. `dcm_volume[:,:,8]`.

Alternatively, you can read one specific slice file right away using its file name, e.g. `dcmread("Data/DICOM_Brain/brain_011.dcm")`.

To display a slice using a python figure, we have to convert it to an pixel array, e.g. `np.array(dcm_slice.pixel_array)`.

You can see the example code below. It might be useful for the following tasks.


```python
# Read a volume and convert to numpy array 
def read_dcm_volume(name):
    volume_img = []
    for file in sorted(os.listdir(name)):
        dcm_slice = dcmread(os.path.join(name, file))
        volume_img.append(dcm_slice.pixel_array)
    return np.transpose(np.array(volume_img),(1,2,0))


# load the volume
dcm_volume = read_dcm_volume('Data/DICOM_Brain')
# extract slice 8 in z-direction
img1 = dcm_volume[:,:,8]


# load one slice
dcm_slice = dcmread("Data/DICOM_Brain/brain_011.dcm")
# convert it to numpy array for plotting the image
img2 = np.array(dcm_slice.pixel_array)


# Create a figure (window)
fig = plt.figure(figsize=(8,4))

# display using gray levels
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('MRI Slice 8')
ax1.imshow(img1, cmap='gray')

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('MRI Slice 11')
ax2.imshow(img2, cmap='gray')
```

**(c)** Did you notice that the sagittal and coronal slice images still seem to be distorted? Working with medical image data is tricky! The *Slice Thickness* you can find in the meta data of a DICOM file is not necessary the actual distance between the slices. You can find further information on this [here](https://stackoverflow.com/a/15116116). To calculate the inter slice spacing between every pair of slices, we can use the variable *Image Position (Patient)*, e.g. using `dicommeta.ImagePositionPatient`, as you can see below:

```python
# Read all dicom files an save them in an array 
def read_all_meta(name):
    dcms = []
    for file in sorted(os.listdir(name)):
        dcm = dcmread(os.path.join(name, file))
        dcms.append(dcm)
    return dcms

# Call the function
meta_data = read_all_meta('Data/DICOM_Brain')

# Create empty array
interSliceDist = np.empty(len(meta_data)-1)

# Calculate the spacing between each slice pair and write it in the array
for i in range(len(meta_data)-1):
    interSliceDist [i] = np.abs(meta_data[i].ImagePositionPatient[2] - meta_data[i-1].ImagePositionPatient[2])

print(interSliceDist)
```

```python
################################################################
# Try out new values for window_center and window_width. 
# Find values that result in an image displaying the skull only.
################################################################

window_center = 400
window_width = 300


## You do not have to change the code below ##

# window function
def window_image(image, window_center, window_width):
    img_min = window_center - window_width // 2
    img_max = window_center + window_width // 2
    window_image = image.copy()
    window_image[window_image < img_min] = img_min
    window_image[window_image > img_max] = img_max
    return window_image

# call the function
window_img =  window_image(dcm_volume[:, :, 10], window_center, window_width)


# Create a figure
fig = plt.figure(figsize=(8,4))

# display using gray levels
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('No Windowing')
ax1.imshow(dcm_volume[:, :, 10], cmap='gray')

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('Windowing')
ax2.imshow(window_img, cmap='gray')
```
